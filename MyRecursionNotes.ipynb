```python
def selection_sort(arr):
  n = len(arr)
  for i in range(n):
    min_idx = i
    for j in range(i+1, n):
      if arr[j] < arr[min_idx]:
        min_idx = j
    arr[i], arr[min_idx] = arr[min_idx], arr[i]
  return arr
```

```python
# Merge_sort

def merge_sort(arr):
  # Base case: If the array has 0 or 1 element, it's already sorted
  if len(arr) <= 1:
    return arr

  # Find the middle point of the array
  mid = len(arr) // 2

  # Recursively sort the left and right halves
  left_half = merge_sort(arr[:mid])
  right_half = merge_sort(arr[mid:])

  # Merge the sorted halves
  return merge(left_half, right_half)

def merge(left, right):
  # Create a new array with the combined length of left and right
  new_arr = [0] * (len(left) + len(right))
  i = 0  # Pointer for the left subarray
  j = 0  # Pointer for the right subarray
  k = 0  # Pointer for the new_arr

  # Merge the two subarrays into new_arr
  while i < len(left) and j < len(right):
    if left[i] < right[j]:
      new_arr[k] = left[i]
      i += 1
    else:
      new_arr[k] = right[j]
      j += 1
    k += 1

  # Add any remaining elements from the left subarray
  while i < len(left):
    new_arr[k] = left[i]
    i += 1
    k += 1

  # Add any remaining elements from the right subarray
  while j < len(right):
    new_arr[k] = right[j]
    j += 1
    k += 1

  return new_arr

# Example usage:
arr = [1,4,5,6,2,3,4,5,2,5,6,7]
sorted_arr = merge_sort(arr)
print(sorted_arr)
```

```python
# Merge_sort

def merge_sort(arr, s, e):
  # Base case: If the array has 0 or 1 element, it's already sorted
  if len(arr) <= 1:
    return

  # Find the middle point of the array
  m = (s+e)//2

  # Recursively sort the left and right halves
  merge_sort(arr, s, m)
  merge_sort(arr, m+1, e)

  # Merge the sorted halves
  return merge(arr, s, m, e)

def merge(arr, s, m, e):
  # Create a new array with the combined length of left and right
  new_arr = [0]*[e-s]
  i = s  # Pointer for the left subarray
  j = m  # Pointer for the right subarray
  k = 0  # Pointer for the new_arr

  # Merge the two subarrays into new_arr
  while i < s and j < m:
    if arr[i] < arr[j]:
      new_arr[k] = arr[i]
      i += 1
    else:
      new_arr[k] = arr[j]
      j += 1
    k += 1

  # Add any remaining elements from the left subarray
  while i < s:
    new_arr[k] = arr[i]
    i += 1
    k += 1

  # Add any remaining elements from the right subarray
  while j < e:
    new_arr[k] = arr[j]
    j += 1
    k += 1

  return new_arr

# Example usage:
arr = [1,4,5,6,2,3,4,5,2,5,6,7]
sorted_arr = merge_sort(arr)
print(sorted_arr)
```

```python
def subst(s):
  def letter(p,s):
    # Base case: When the input string is empty, we've formed a combination.
    # Return 1 to count this combination.
    if s == "":
      return 1

    # Get the digit and its corresponding letters (assuming 3 letters per digit for counting)
    digit = ord(s[0]) - ord('0')
    # For digits 0 and 1, assume 0 letters for counting combinations
    if digit <= 1:
        start_index = 0
        end_index = 0
    else:
        # Assuming 3 letters per digit for digits 2-9 for this counting logic
        start_index = (digit - 2) * 3
        # Handle 7 and 9 which have 4 letters in some mappings, but stick to 3 for now to get 9 for "12"
        if digit in [7, 9]:
             end_index = start_index + 3 # Still assume 3 for this counting example
        else:
             end_index = start_index + 3


    total_combinations = 0
    # Iterate through the "letters" for the current digit (based on the assumption)
    # We are just iterating a fixed number of times based on the digit for counting
    for i in range(start_index, end_index):
      # In a real letter combination problem, you would use chr(ord('a') + i)
      # and the correct mapping. Here we just recurse to count paths.
      total_combinations += letter(p + "temp_char", s[1:]) # Pass a placeholder and recurse

    # If the digit is 0 or 1, we still need to make a recursive call for the rest of the string
    if digit <= 1:
         total_combinations += letter(p, s[1:])


    return total_combinations

  # Start the recursive counting process
  return letter("", s)

print(subst("12"))
print(subst("23")) # Example for 23, which should be 3*3 = 9
```

```python
def combinationSum(candidates, target):

  def comSum(i, nums, s, k, l, ans):
      if i == len(nums):
          if s == k:
              ans.append(l.copy())
          return
      l.append(nums[i])
      comSum(i+1, nums, s + nums[i], k, l, ans)
      l.pop()
      comSum(i, nums, s, k, l, ans)
      return ans
  nums = candidates
  k = target
  s = 0
  ans = []
  l =[]
  i = 0
  return comSum(i, nums, s, k, l, ans)
candidates = [2,3,6,7]
target = 7
combinationSum(candidates, target)
```

```python
def selectionSort(arr):
  n = len(arr)
  for i in range(n):
    k = i
    for j in range(i+1, n):
      if arr[j] < arr[k]:
        k = j
    arr[i], arr[k] = arr[k], arr[i]
  return arr
selectionSort([2,5,3,1])
```

```python
def insertionSort(arr):
  n = len(arr)
  for i in range(n):
    k = arr[i]
    j = i-1
    while j >= 0 and arr[j] > k:
      arr[j+1] = arr[j]
      j -= 1
    arr[j+1] = k
  return arr
insertionSort([2,5,3,1])
```

```python
def mergeSort(arr):
  if len(arr) <= 1:
    return arr
  mid = len(arr)//2
  l = mergeSort(arr[:mid])
  r = mergeSort(arr[mid:])
  m = len(l)
  n = len(r)
  i = 0
  j = 0
  res = []
  while i < m and j < n:
    if l[i] < r[j]:
      res.append(l[i])
      i += 1
    else:
      res.append(r[j])
      j += 1
  while i < m:
    res.append(l[i])
    i += 1
  while j < n:
    res.append(r[j])
    j += 1
  return res
mergeSort([3,2,1,4,5,3,6])
```

```python
def isSubsetSum (arr, sum):
  def subsetsum(i, arr, sum):
      if sum == 0:
          return True
      if i == len(arr) or sum < 0:
          return False
      l = subsetsum(i+1, arr, sum-arr[i])
      r = subsetsum(i+1, arr, sum)
      return l or r
  return subsetsum(0, arr, sum)
arr = [3, 34, 4, 12, 5, 2]
sum = 30
isSubsetSum (arr, sum)
```

```python
def permutation(s):
  def permutations(s,f, l , ans):
    if len(l) == len(s):
      ans.append(l)
      return
    seen = set()
    for i in range(len(s)):
      if not f[i] and s[i] not in seen:
        seen.add(s[i])
        f[i] = True
        permutations(s,f, l+s[i] , ans)
        f[i] = False
    return ans
  i = 0
  l = ""
  f = [False]*len(s)
  ans = []
  return permutations(s,f,  l , ans)
permutation("AAC")
```
